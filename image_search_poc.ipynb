# Поиск референсных фотографий по описанию.

Данный проект направлен на создание нового сервиса на платформе фотохостинга для профессиональных фотографов «Со Смыслом» (With Sense) - поиска референсных фотографий по запросу. Суть сервиса: пользователь вводит текстовое описание сцены, и система возвращает фотографии с подобными сценами.

Цель проекта — разработать и продемонстрировать Proof of Concept (PoC) поиска изображений по текстовым описаниям, подтвердив тем самым возможность практической реализации этого сервиса.

В проекте будет осуществляться решение задачи регрессии с применением методов обучения с учителем.

**Описание данных**

Для обучения и тестирования моделей используются следующие данные:

`train_dataset.csv` - содержит информацию, необходимую для обучения: имена файлов изображений, идентификаторы описаний и тексты описаний.

`CrowdAnnotations.tsv` - данные по соответствию изображения и описания, собранные через краудсорсинг. Номера колонок и соответствующий тип данных:
1.	Имя файла изображения.
2.	Идентификатор описания.
3.	Доля людей, подтвердивших, что описание соответствует изображению.
4.	Количество человек, подтвердивших, что описание соответствует изображению.
5.	Количество человек, подтвердивших, что описание не соответствует изображению.
	
`ExpertAnnotations.tsv` - данные по соответствию изображения и описания, оценённые экспертами. Номера колонок и соответствующий тип данных:
1.	Имя файла изображения.
2.	Идентификатор описания.
3, 4, 5 — оценки трёх экспертов. (1 — изображение и запрос совершенно не соответствуют друг другу, 2 — запрос содержит элементы описания изображения, но в целом запрос тексту не соответствует, 3 — запрос и текст соответствуют с точностью до некоторых деталей, 4 — запрос и текст соответствуют полностью)

`test_queries.csv` - содержит информацию, необходимую для тестирования: идентификатор запроса, текст запроса и релевантное изображение.

`test_images.csv` - содержит список имен файлов изображений для тестирования.

`train_images` и `test_images`: папки с изображениями для обучения и тестирования.

Количество наблюдений в таблицах с обучающими данными и оценками экспертов одинаково.Это позволяет предположить, что для каждой пары  "изображение - описание" есть экспертная оценка. Проверим так ли это.

```python
def img_text_check(df_1, df_2):
    # Создание множеств пар (image, query_id) для обеих таблиц
    first_pairs = set(zip(df_1['image'], df_1['query_id']))
    second_pairs = set(zip(df_2['image'], df_2['query_id']))

    # Проверка, все ли пары из train_pairs присутствуют в expert_pairs
    all_pairs_present = first_pairs.issubset(second_pairs)

    # Вывод результата
    if all_pairs_present:
        print('Для всех пар "изображение - описание" есть оценка.')
    else:
        print('Не для всех пар "изображение - описание" есть оценка.')
        print(f"Количество отсутствующих пар: {len(first_pairs - second_pairs)}")
```

```python
def missing_img(df, path):
    # получение уникальных значений в столбце 'image'
    unique_images_from_df = df['image'].unique()

    # получение списока всех файлов в папке с фото
    image_files_in_folder = set(os.listdir(path))

    # проверка наличия изображения из столбца 'image' в папке с фото
    missing_files = [image for image in unique_images_from_df if image not in image_files_in_folder]

    if not missing_files:
        print("В папке с изображениями отсутствующих файлов не обнаружено")
    else:
        print("В папке с изображениями не обнаружено следующих файлов:")
        print(missing_files)
```

Объявим класс-обертку для датасета, который будет возвращать пути к изображениям.

```python
class ImagePathDataset(Dataset): # Определяем новый класс ImagePathDataset, который наследуется от Dataset.
    def __init__(self, file_names, img_dir, transform=None):
        self.file_names = file_names  # Массив с именами файлов
        self.img_dir = img_dir  # Директория с изображениями
        self.transform = transform  # Преобразования для изображений

    def __len__(self):
        return len(self.file_names)  # Количество изображений

    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, self.file_names[idx])  # Полный путь к изображению
        image = Image.open(img_name).convert('RGB')  # Загружаем и конвертируем изображение в RGB формат
        
        if self.transform:
            image = self.transform(image)  # Применяем преобразования, если они заданы
        
        return image, self.file_names[idx]  # Возвращаем изображение и его имя
```

Объявим функцию векторизации изображения.

```python
def img_to_vec(model, image_loader):
    '''
    Принимает на вход обученную модель и объект DataLoader для набора данных с изображениями.
    Возвращает таблицу с векторами изображений и именами файлов.
    '''
    model = model.to(device)  # Перенос модели на GPU
    model.eval() # Включаем режим предсказаний
    
    # Извлечение признаков для всех изображений
    image_vectors = []
    image_names = []
    
    with torch.no_grad():
        for images, img_names in image_loader:
            images = images.to(device)  # Перенос изображений на GPU
            outputs = model(images)
            outputs = outputs.view(outputs.size(0), -1)  # Преобразование вектора в двумерный тензор
            outputs = outputs.cpu()  # Перенос выходных данных обратно на CPU для дальнейшей обработки
            for output, img_name in zip(outputs, img_names):
                image_vectors.append(output.numpy())
                image_names.append(img_name)

    # Создание DataFrame с векторами признаков
    image_vector_df = pd.DataFrame(image_vectors)
    image_vector_df['image'] = image_names
    image_vector_df = image_vector_df.rename(columns=lambda x: 'img_' + str(x) if x != 'image' else x)
    
    model = model.cpu() # Перенос модели обратно на CPU
    
    return image_vector_df
```

### Агрегация оценок.

Оценка соответствия текста изображению - это целевая переменная для моделей. Она должна иметь значения от 0 до 1. 

Для агрегации оценок экспертов просуммируем проставленные ими баллы и вычислим, какую долю от максимально возможной оценки они составляют, при этом, если все 3 специалиста сошлись во мнении, что "изображение и запрос совершенно не соответствуют друг другу" оценка должна быть равной 0.

Объявим функцию агрегации экспертных оценок

```python
def experts_score(row):
    x_1, x_2, x_3 = row['score_1'], row['score_2'], row['score_3']
    score = (x_1 + x_2 + x_3) / 12
    if score > 0.25:
        return score
    else:
        return 0
```

Качество предсказаний моделей будет оцениваться с помощью метрики RMSE (Root Mean Squared Error, среднеквадратическая ошибка). Эта метрика более чувствительна к большим ошибкам по сравнению, например, с MAE (Mean Absolute Error). Кроме того, линейные модели и нейронные сети лучше обучаются с функцией потерь, основанной на квадрате ошибок. Поэтому использование RMSE как метрики качества позволяет напрямую соотноситься с функцией потерь, применяемой при обучении модели.

Создадим простую нейронную сеть для задачи регрессии.

```python
class RegressionModel(pl.LightningModule):
    def __init__(self, input_dim, hidden_dim1, hidden_dim2, hidden_dim3, learning_rate):
        super(RegressionModel, self).__init__()
        self.layer_1 = torch.nn.Linear(input_dim, hidden_dim1)
        self.layer_2 = torch.nn.Linear(hidden_dim1, hidden_dim2)
        self.layer_3 = torch.nn.Linear(hidden_dim2, hidden_dim3)
        self.output = torch.nn.Linear(hidden_dim3, 1)
        self.learning_rate = learning_rate
        self.rmse = torchmetrics.MeanSquaredError(squared=False)

    def forward(self, x):
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        x = F.relu(self.layer_3(x))
        x = self.output(x)
        x = F.sigmoid(x)
        return x

    def training_step(self, batch, batch_idx):
        x, y = batch
        y = y.unsqueeze(1)
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y = y.unsqueeze(1)
        y_hat = self(x)
        val_loss = F.mse_loss(y_hat, y)
        rmse = self.rmse(y_hat, y)
        self.log('val_loss', val_loss)
        self.log('val_rmse', rmse)
        return val_loss
    
    def predict_step(self, batch, batch_idx, dataloader_idx=0):
        return self(batch)
        
    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.learning_rate)
        return optimizer
```

```python
class PrintMetricsCallback(Callback):
    def on_validation_end(self, trainer, pl_module):
        # Выводим логи после каждой валидационной эпохи
        val_loss = trainer.callback_metrics.get('val_loss')
        val_rmse = trainer.callback_metrics.get('val_rmse')
        
        print(f'Epoch {trainer.current_epoch}: val_loss={val_loss}, val_rmse={val_rmse}')
```

```python
def net_objective(trial):
    architectures = [
        (1024, 512, 256),
        (1024, 256, 64),
        (512, 256, 128),
        (512, 64, 32)
    ]
    
    hidden_dims = trial.suggest_categorical('hidden_dims', architectures)
    hidden_dim1, hidden_dim2, hidden_dim3 = hidden_dims

    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2)

    model = RegressionModel(
        input_dim=x_train_tensor.shape[1],
        hidden_dim1=hidden_dim1,
        hidden_dim2=hidden_dim2,
        hidden_dim3=hidden_dim3,
        learning_rate=learning_rate
    )

    trainer = Trainer(
        max_epochs=10,
        logger=TensorBoardLogger("lightning_logs", name="optuna"),
        gpus=1
    )

    trainer.fit(model, train_loader, valid_loader)
    return trainer.callback_metrics['val_rmse'].item()
```

### Вывод наиболее релевантного изображения для каждого запроса и визуальный анализ качетсва поиска.

Объявим функцию для поиска изображений соответствующим текстовому описанию с наибольшей вероятностью. Если запрос ведёт на юридически вредный контент, функция выведет дисклеймер: "This image is unavailable in your country in compliance with local laws."

```python
def find_img(text, image_df, model, path_to_img_folder, batch_size):
    # Лемматизация текста и удаление стоп-слов
    text_lemma = do_lemma_del_stop(text)
    
    # Кодирование текста в вектор
    txt_vector = encoder.encode(text_lemma)
    
    # Получение векторов изображений
    img_vectors = image_df.drop(columns='image').values
    
    # Создание массива векторов текста, соответствующих каждому изображению
    txt_vector_resized = np.tile(txt_vector, (img_vectors.shape[0], 1))
    
    # Объединение векторов изображений и текста
    data = np.concatenate((img_vectors, txt_vector_resized), axis=1)
    
    # Преобразование данных в тензор
    tensor = torch.FloatTensor(data)
    
    # Создание DataLoader
    loader = DataLoader(tensor, batch_size=batch_size, shuffle=False)
    
    # Получение предсказаний модели
    preds = trainer.predict(model, loader)
    
    # Преобразование списка тензоров в один тензор
    all_preds_tensor = torch.cat(preds)
    
    # Преобразование тензора в numpy массив
    all_preds_array = all_preds_tensor.numpy()
    
    # Нахождение индекса максимального значения
    max_index = np.argmax(all_preds_array)
    
    # Получение имени изображения по найденному индексу
    image = image_df['image'].iloc[max_index]
    
    # Формирование пути к изображению
    image_path = f"{path_to_img_folder}/{image}"
    
    try:
        # Открытие изображения
        img = Image.open(image_path).convert('RGB')
        
        # Вывод информации 
        print('\033[1m' + f'Текст запроса:' + '\033[0m' + f' {text}')
        print(f'Вероятность соответствия текста изображению: {all_preds_array[max_index].item() * 100:.2f}%')
        
        # Отображение изображения
        plt.imshow(img)
        plt.axis('off')
        plt.show()
        
    except FileNotFoundError:
        print(f'Изображение не найдено по пути: {image_path}')
    except Exception as e:
        print(f'Ошибка при открытии изображения: {e}')
        
    return None
```

Объявим функцию для предварительной проверки на "вредный контент".

```python
def fin_legal_image(text, image_df, model, path_to_img_folder='data/test_images/', batch_size=32):
    if contains_restricted_word(text):
        print('\033[1m' + f'Текст запроса:' + '\033[0m' + f' {text}')
        print('\033[31m' + f'This image is unavailable in your country in compliance with local laws.' + '\033[0m')
        print()
        return None
    else:
        find_img(text, image_df, model, path_to_img_folder, batch_size)
```
